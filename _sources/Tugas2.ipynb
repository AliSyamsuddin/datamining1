"missing value" atau "nilai yang hilang" merujuk pada keadaan di mana tidak ada nilai yang diisi atau tersedia untuk suatu variabel dalam dataset. Missing value bisa terjadi karena berbagai alasan, termasuk kesalahan pengumpulan data, kegagalan sensor, atau karena data tersebut tidak relevan atau tidak tersedia pada saat pengumpulan.

Ketika nilai yang hilang terjadi dalam dataset, hal itu dapat mempengaruhi analisis data dan hasil yang dihasilkan. Beberapa dampaknya antara lain:

1. Penurunan Akurasi: Jika missing value tidak ditangani dengan baik, dapat mengurangi akurasi analisis dan model yang dibangun dari dataset tersebut.

2. Bias: Kehadiran missing value dapat menyebabkan bias dalam analisis data, karena penghapusan atau pengisian nilai yang hilang dapat mempengaruhi distribusi data dan hasil analisis.

3. Pengurangan Informasi: Missing value dapat mengurangi jumlah informasi yang tersedia untuk analisis, yang dapat mengurangi kegunaan dataset dan mengurangi kekuatan analisis.

Untuk menangani missing value, beberapa strategi umum yang sering digunakan adalah:

- Penghapusan Data: Menghapus baris atau kolom yang mengandung nilai yang hilang. Namun, ini dapat mengurangi jumlah data yang tersedia untuk analisis.

- Imputasi: Mengisi nilai yang hilang dengan nilai yang diestimasi, seperti rata-rata, median, atau modus dari data yang ada. Imputasi harus dilakukan dengan hati-hati untuk mempertahankan integritas data.

- Metode yang Lebih Kompleks: Terkadang, metode imputasi yang lebih kompleks seperti regresi, pengisian nilai berdasarkan pola atau hubungan dengan variabel lain, atau menggunakan teknik seperti k-Nearest Neighbors (KNN) dapat digunakan tergantung pada jenis data dan konteks analisis.

Pemilihan strategi untuk menangani missing value harus dipertimbangkan dengan cermat berdasarkan jenis data, jumlah missing value, dan tujuan analisis data yang dilakukan.

# ğŸš. ğ•„ğ•–ğ•¥ğ• ğ••ğ•– ğ•‚â„•â„•

Metode KNN (K-Nearest Neighbors) adalah sebuah algoritma pembelajaran mesin yang digunakan dalam berbagai jenis tugas, termasuk klasifikasi dan regresi. Namun, di sini kita akan fokus pada penggunaannya dalam konteks penanganan nilai yang hilang (missing value) dalam dataset.

Dalam konteks penanganan missing value, metode KNN digunakan untuk mengisi nilai yang hilang dengan memperkirakan nilai yang paling cocok berdasarkan data yang ada. Secara khusus, langkah-langkah umum yang dilakukan dalam menggunakan metode KNN untuk penanganan missing value adalah sebagai berikut:

## Langkah-Langkah Imputasi

1. Pertama, kita tentukan nilai k (jumlah tetangga terdekat yang akan digunakan).
2. Kemudian, untuk setiap titik data yang memiliki nilai yang hilang, kita cari k tetangga terdekat dari titik tersebut berdasarkan atribut-atribut yang relevan.
3. Selanjutnya, kita hitung rata-rata atau median dari nilai-nilai atribut yang relevan dari tetangga-tetangga ini.
4. Nilai tersebut kemudian digunakan untuk mengisi nilai yang hilang pada titik data tersebut.

Pemilihan Jarak: Dalam KNN, pemilihan jarak atau metrik jarak adalah faktor penting. Anda dapat menggunakan berbagai metrik jarak, seperti jarak Euclidean, Manhattan, atau lainnya, tergantung pada jenis data dan distribusi variabel.

Pemilihan K: Pemilihan nilai k juga penting. K yang terlalu kecil dapat menyebabkan hasil yang tidak stabil, sedangkan K yang terlalu besar dapat menyebabkan pemusatan yang berlebihan pada nilai rata-rata.

## Local Outlier Factor

Secara singkat, Local Outlier Factor (LOF) adalah metode deteksi outlier yang mempertimbangkan kerapatan titik data dalam lingkungan lokal mereka untuk menentukan sejauh mana titik-titik tersebut berbeda dari tetangga-tetangganya. LOF memberikan skor pada setiap titik data, menyoroti titik-titik yang memiliki kerapatan yang rendah dibandingkan dengan lingkungan lokalnya, dan kemungkinan besar merupakan outlier. Metode ini berguna untuk mendeteksi outlier dalam dataset yang besar dan kompleks. Beberapa konsep LOF adalah sebagai berikut :

1. K-Distance (Jarak K):
K-Distance (jarak K) adalah jarak dari sebuah titik data ke titik data terdekat ke-K dalam suatu dataset. Dalam konteks deteksi outlier, jarak K biasanya digunakan dalam perhitungan Local Reachability Density (LRD) dalam algoritma LOF. Ini membantu dalam menentukan kerapatan lokal dari titik data dengan memperhitungkan seberapa jauh titik tersebut dari tetangga-tetangganya.

2. K-Neighbors (Tetangga K):
K-Neighbors (tetangga K) adalah K titik data terdekat dengan sebuah titik data tertentu dalam suatu dataset. Dalam algoritma K-Nearest Neighbors (KNN), K-Neighbors digunakan untuk membuat prediksi atau mengklasifikasikan titik data berdasarkan mayoritas kelas dari tetangga-tetangganya. Dalam konteks LOF, K-Neighbors digunakan untuk menghitung Local Reachability Density (LRD) dari titik data, yang digunakan dalam perhitungan LOF skor.

# ğŸ›. ğ•ƒğ• ğ•”ğ•’ğ• â„ğ•–ğ•’ğ•”ğ•™ğ•’ğ•“ğ•šğ•ğ•šğ•¥ğ•ª ğ”»ğ•–ğ•Ÿğ•¤ğ•šğ•¥ğ•ª (ğ•ƒâ„ğ”»)

Local Reachability Density (LRD) adalah ukuran kepadatan lokal dari sebuah titik data dalam dataset. Ini digunakan dalam algoritma Local Outlier Factor (LOF) untuk mengevaluasi seberapa dekat atau terisolasi titik tersebut dari lingkungan tetangganya. LRD tinggi menunjukkan titik data yang berada dalam lingkungan yang padat, sedangkan LRD rendah menunjukkan isolasi relatif dari tetangga-tetangganya. Ini membantu dalam identifikasi outlier dalam analisis data.

<img src="image-20240320-220736.png" width="" align="" />

Di sini, semakin tinggi nilai LRD sebuah titik data, semakin padat atau dekat titik data tersebut dengan tetangga-tetangganya. LRD yang tinggi menandakan bahwa titik data tersebut berada dalam wilayah dengan kepadatan tinggi, sedangkan LRD yang rendah menunjukkan bahwa titik data tersebut berada dalam wilayah dengan kepadatan rendah. Dalam hal ini, x adalah titik data yang sedang dinilai, y adalah salah satu dari K tetangga terdekat dari x, dan RD(x,y) adalah Reachability Distance antara x dan y.

# ğŸœ. ğ•ƒğ• ğ•”ğ•’ğ• ğ•†ğ•¦ğ•¥ğ•ğ•šğ•–ğ•£ ğ”½ğ•’ğ•”ğ•¥ğ• ğ•£ (ğ•ƒğ•†ğ”½)

Local Outlier Factor (LOF) adalah metode deteksi outlier yang digunakan dalam analisis data. LOF mengevaluasi seberapa anomali atau "outlying" sebuah titik data dalam dataset dengan mempertimbangkan lingkungan lokalnya. Secara khusus, LOF membandingkan kerapatan titik data tersebut dengan kerapatan tetangga-tetangganya. Jika titik data memiliki kerapatan yang lebih rendah dari tetangga-tetangganya, maka nilai LOF-nya akan tinggi, menandakan bahwa titik tersebut mungkin merupakan outlier. LOF adalah alat yang berguna untuk mengidentifikasi titik-titik data yang berbeda dari lingkungan sekitarnya dalam dataset.

<img src="image-20240320-221121.png" width="" align="" />

LRD tiap titik digunakan untuk membandingkan dengan rata-rata LRD K tetangganya. LOF adalah perbandingan rata-rata LRD K tetangga dari sebuah titik A dengan LRD A.

Jika suatu titik bukan merupakan pencilan (inlier), rasio rata-rata LRD tetangganya hampir sama dengan LRD titik itu sendiri (karena kepadatan titik dan tetangganya hampir sama). Dalam kasus ini, LOF akan mendekati nilai 1. Sebaliknya, jika suatu titik merupakan outlier, LRD titik tersebut lebih kecil dari rata-rata LRD tetangganya. Akibatnya, nilai LOF akan tinggi.

Meskipun LOF > 1 seringkali menandakan outlier, hal ini tidak selalu benar. Misalkan jika kita mengetahui bahwa hanya ada satu outlier dalam data, kita dapat mengambil nilai LOF maksimum di antara semua nilai LOF, dan titik yang sesuai dengan nilai LOF maksimum akan dianggap sebagai outlier.

```
import pandas as pd
import mysql.connector
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# koneksi ke database MySQL
conn = mysql.connector.connect(
    host="database1-alisyamsuddin22086.a.aivencloud.com",
    user="avnadmin",
    password="AVNS_zWY_hBUclXzcicAmP1r",
    port = "22753",
    database="tugas1"
)

# Membaca data dari database menggunakan Pandas
sql_query = "SELECT * FROM tableName"
data_iris = pd.read_sql(sql_query, conn)

# Menghapus kolom Id, dikarenakan kurang relevan
data_iris.drop('Id',axis=1,inplace=True)

# memisahkan atribut dan label
X = data_iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm' ]]
y = data_iris['Species']
 
# Membagi dataset menjadi data latih & data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
 
# membuat model Gaussian naive bayes
gnb_model = GaussianNB()
 
# Melatih model dengan menggunakan data latih
gnb_model = gnb_model.fit(X_train, y_train)

# Evaluasi Model
y_pred = gnb_model.predict(X_test)
 
acc_secore = round(accuracy_score(y_pred, y_test), 3)
print('Accuracy: ', acc_secore)
```
